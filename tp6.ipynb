## Fonction binomiale



1) Écrire une fonction `factorielle` qui renvoie la valeur factorielle de `n` où `n` étant passé en paramètre. Par exemple $1!=1$ et $5!=120$. On convient que $0!=1$. On écrira une version de la fonction factorielle utilisant une boucle `for`.



```python
def factorielle(n):
    fac = 1
    for cpt in range(2, n + 1):
        fac *= cpt
    return fac

for _ in range(3):
    print(factorielle(int(input())), end = "\n\n")
```

2) On rappelle que si $n$ et $p$ sont des entiers positifs ou nuls, le coefficient binomial est défini par

$$\binom{n}{p}=\begin{cases}\displaystyle\frac{n!}{p!(n-p)!}& \text{si } 0\leq p\leq n\\0& \text{sinon}\end{cases}$$

Ecrire une fonction `binomial(n, p)` qui renvoie le coefficient binomial ci-dessus.



```python
def factorielle(n):
    fac = 1
    for cpt in range(2, n + 1):
        fac *= cpt
    return fac

def binomial(n, p):
    if 0 <= p <= n: return factorielle(n)//(factorielle(p)*factorielle(n - p))
    return 0
```

3) Ecrire une fonction `pascal` qui vérifie l'identité  de Pascal : $\binom{n}{p}+\binom{n}{p+1}=\binom{n+1}{p+1}$ valable pour tous entiers $n, p\geq 0$. Vérifier l'identité pour tous les entiers $n$ et $p$ entre 0 et 100.



```python
def factorielle(n):
    fac = 1
    for cpt in range(2, n + 1):
        fac *= cpt
    return fac

def binomial(n, p):
    if 0 <= p <= n: return factorielle(n)//(factorielle(p)*factorielle(n - p))
    return 0

def pascal(n, p):
    return binomial(n, p) + binomial(n, p + 1) == binomial(n + 1, p + 1)

for entN in range(101):
    for entP in range(101):
        if not pascal(entN, entP): print(entN, entP)
        print(pascal(entN, entP))
```

4) Ecrire une fonction `maxBinomial(n)` qui renvoie la valeur maximale de tous les coefficients binomiaux  $\binom{n}{0}, \binom{n}{1}, \binom{n}{2}, \cdots , \binom{n}{n}$



```python
def factorielle(n):
    fac = 1
    for cpt in range(2, n + 1):
        fac *= cpt
    return fac

def binomial(n, p):
    if 0 <= p <= n: return factorielle(n)//(factorielle(p)*factorielle(n - p))
    return 0

def maxBinomial(n):
    return max([binomial(n, p) for p in range(n + 1)])
```

5) Ecrire une fonction `verifMaxBinomial(n)` qui vérifie que la valeur maximale de tous les coefficients binomiaux $\binom{n}{0}, \binom{n}{1}, \binom{n}{2}, \cdots, \binom{n}{n}$ est $\binom{n}{m}$ où $m$ est le quotient la division entière de $n$ par 2.

## Disques creux



1) Écrire une procédure `disque` permettant d'afficher avec Turtle un disque creux dont les coordonnées `x` et `y` du centre et le diamètre `d` sont passés en paramètre. On tracera un `dot` de couleur `white` à l'intérieur d'un `dot` de couleur `black`.



```python
from turtle import *
reset()
ht()

def disque(x, y, d):
    up()
    goto(x, y)
    down()
    dot(d, "black")
    dot(d//2, "white")

disque(100, 100, 100)
```

2) Dans le programme principal, dessiner une grille de disques ayant la forme de la figure suivante.

![](disques_trou.png)

L'espacement entre les disques correspond au diamètre de chaque disque noir.



```python
from turtle import *
reset()

def disque(x, y, d):
    up()
    goto(x, y)
    down()
    dot(d, "black")
    dot(d//2, "white")

def grille(nb, d):
    start = -(nb - 1)*d//2
    y = -start
    for _ in range(nb):
        x = start
        for _ in range(nb):
            disque(x, y, d)
            x += d
        y -= d

grille(int(input()), int(input()))
```
